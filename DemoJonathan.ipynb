# *Visualisation rapide de notre BDD* #

```python
import pandas as pd
df=pd.read_csv('Data/promo.csv', index_col=0) #index col pour ne pas mettre une colonne d'index supplémentaire par défault
df.head()
```

# *Analyse du rapport Posteur de Gifs / Nombre de Sucettes* #

```python
import seaborn as sns

sns.lmplot(x='Sucettes', y='Gifs', data=df) #Fonctionne avec deux données chiffrées

#lmplot : bien pour modèles de regression
```

```python
# Scatterplot arguments
sns.lmplot(x='Sucettes', y='Gifs', data=df,
           hue='Fonction')   # Color by evolution stage
#tester hue=GoldSuc
```

*a general rule is that it makes sense to use hue for the most important comparison, followed by col and row. However, always think about your particular dataset and the goals of the visualization you are creating.*

*On enrichie avec des méthodes Matplotlib*

```python
import matplotlib.pyplot as plt #Sans le pyplot on a pas les attributs ylim xlim, simplifie fonctions#

# Plot using Seaborn
sns.lmplot(x='Sucettes', y='Gifs', data=df,
           fit_reg=False, # No regression line
           hue='Fonction',
           col='Genre')
 
# Tweak using Matplotlib
plt.ylim(0, None) #permet de sélectionner l'abcisse minimum max et ordonnée min max pour voir une partie précise
plt.xlim(0, None)
```

# *Débit / Ponctualité* #

```python
# Distribution Plot (a.k.a. Histogram)
sns.displot(df.Débit) 
```

```python
# Distribution Plot (a.k.a. Histogram)
sns.displot(x='Sucettes', y='Débit', data=df, palette=("Blues_d")) #pas top en remplacant count par attribut, 
#histogramme moins de sens
```

```python
# Distribution Plot (a.k.a. Histogram)
sns.displot(x='Débit', data=df, kind="kde") #kind=mode kde=densité  kernel density estimates
```

```python
sns.displot(data=df, x="Débit", hue="Simplonien", col="Ponctualité", kind="ecdf") #empirical cumulative distribution functions

# Tweak using Matplotlib
plt.ylim(0, None) 
plt.xlim(0, None)
```

On remarque que ceux qui ont le moins de débit sont moins ponctuel.

```python
sns.displot(data=df, x="Gifs", kind="ecdf")
```

# *Boisson / Sucettes* #

```python
# Variable palette de couleurs attribuées aux Boissons
PaletteBoisson = ['#8B4513',  # Café
                    '#98D8D8',  # Eau
                    '#E0C068',  # Thé
                    '#F08030',  # Rhum
                   ]
```

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Boisson', data=df, palette=PaletteBoisson) 
 
# Rotate x-labels
plt.xticks(rotation=-45)

#Show the counts of observations in each categorical bin using bars.
```

```python

# Count Plot (a.k.a. Bar Plot)

sns.set_theme(style="darkgrid")
sns.countplot(x='Boisson', data=df, hue='Fonction', palette=PaletteBoisson) #Test y= et inverser y et hue
 
# Rotate x-labels
plt.xticks(rotation=-45)

#Show the counts of observations in each categorical bin using bars.
```

```python
plt.figure(figsize = [14,9])
sns.set_theme(style="darkgrid")
sns.countplot(x='Simplonien', data=df, hue='Boisson', palette=PaletteBoisson) #Test y= et inverser y et hue



 
# Rotate x-labels
plt.xticks(rotation=-45)

#Show the counts of observations in each categorical bin using bars.
```

```python
# catplot, anciennement Factor Plot
g = sns.catplot(x='Pratique', 
                   y='Sucettes', 
                   data=df,
                   hue='Simplonien',  # Color by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
 
# Doesn't work because only rotates last plot
# plt.xticks(rotation=-45)

#This function provides access to several axes-level functions that show the relationship between a numerical and one
#or more categorical variables using one of several visual representations. The kind parameter selects the underlying
#axes-level function to use
```

```python
# Joint Distribution Plot
sns.jointplot(x='Sucettes', y='GoldSuc', hue='Genre', data=df, kind='kde')



#Draw a plot of two variables with bivariate and univariate graphs.
```
